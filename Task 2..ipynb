```python
data1 = data[data['Species']=='Iris-setosa']
data2 = data[data['Species']=='Iris-versicolor']
data3 = data[data['Species']=='Iris-virginica']
```

```python
plt.scatter(data1['PetalLengthCm'],data1['PetalWidthCm'], color='r' , label='Iris-setosa')
plt.scatter(data2['PetalLengthCm'],data2['PetalWidthCm'], color='b', label='Iris-versicolor')
plt.scatter(data3['PetalLengthCm'],data3['PetalWidthCm'], color='g' , label='Iris-virginica ')


plt.legend()
plt.show()
```

```python
data_imp = data.iloc[:,0:4]
from sklearn.cluster import KMeans
k_meansclus = range(1,10)
sse = []

for k in k_meansclus :
  km = KMeans(n_clusters =k)
  km.fit(data_imp)
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.title('The Elbow Method')
plt.plot(k_meansclus,sse)
plt.show()
```

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(data_imp)
```
